```python
#Linear regression example
import numpy as np
import matplotlib.pyplot as plt
# Given data
heights = np.array([60, 62, 64, 66, 68])
actual_weights = np.array([115, 120, 130, 140, 150])

# Step 1: Compute mean of X and Y
mean_heights = np.mean(heights)
mean_weights = np.mean(actual_weights)

# Step 2: Compute the deviations and their squares
deviation_heights = heights - mean_heights
deviation_weights = actual_weights - mean_weights

# Step 3: Compute the cross-deviation and its square
cross_deviation = np.sum(deviation_heights * deviation_weights)

# Step 4: Compute the squared deviations of X
squared_deviation_heights = np.sum(deviation_heights ** 2)

# Step 5: Compute the slope (m)
m = cross_deviation / squared_deviation_heights

# Step 6: Compute the intercept (c)
c = mean_weights - m * mean_heights

# Step 7: Predicted weights
predicted_weights = m * heights + c

# Visualization
plt.scatter(heights, actual_weights, color='blue', label='Actual')
plt.plot(heights, predicted_weights, color='red', label='Predicted')
plt.title('Linear Regression')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend()
plt.show()

# Print the slope and intercept
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
ols=np.sum((predicted_weights-actual_weights)**2)
print(ols)
```

```python
sme=np.sum(abs((predicted_weights-actual_weights)))/len(heights)
print(sme)
```

```python
meansquareerror=np.sum((predicted_weights-actual_weights)**2)/len(heights)
print(meansquareerror)
```

```python
import math
rmse=math.sqrt(meansquareerror)
print(rmse)
```
