```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def calculate_slope(x, y):
 
 
  mean_x = np.mean(x)
  mean_y = np.mean(y)

 
  covariance = np.sum((x - mean_x) * (y - mean_y)) / (len(x) - 1)
  variance = np.sum((x - mean_x) ** 2) / (len(x) - 1)


  slope = covariance / variance
  return slope
```

```python

np.random.seed(0)
study_hours = np.random.uniform(0, 10, 50)  # Random study hours
exam_scores = 2 * study_hours + np.random.normal(0, 2, 50)  # Exam scores with noise
# Introduce an outlier
study_hours[0] = 1
exam_scores[0] = 90

# Scatter plot with regression line before outlier removal
plt.figure(figsize=(10, 6))
plt.scatter(study_hours, exam_scores, label='Data')
model = LinearRegression()
model.fit(study_hours.reshape(-1, 1), exam_scores)
plt.plot(study_hours, model.predict(study_hours.reshape(-1, 1)), color='red', label='Regression Line')
plt.title('Scatter Plot with Regression Line (Before Outlier Removal)')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.legend()
plt.show()

# Remove outlier
filtered_study_hours = np.delete(study_hours, 0)
filtered_exam_scores = np.delete(exam_scores, 0)

# Scatter plot with regression line after outlier removal
plt.figure(figsize=(10, 6))
plt.scatter(filtered_study_hours, filtered_exam_scores, label='Data (Outlier Removed)')
model = LinearRegression()
model.fit(filtered_study_hours.reshape(-1, 1), filtered_exam_scores)
plt.plot(filtered_study_hours, model.predict(filtered_study_hours.reshape(-1, 1)), color='green', label='Regression Line (Outlier Removed)')
plt.title('Scatter Plot with Regression Line (After Outlier Removal)')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')

plt.legend()
plt.show()
```
